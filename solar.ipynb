```python
# Plot Distribution of Generated Power Ussing Histogram
plt.Figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'], bins = 30,kde=True)
plt.title("Generated Power Distribution")
plt.xlabel("Generated power in KW")
plt.ylabel("Frequency")
plt.show()
```

```python
# using scatter plot to analyse cloud cover with respective generated power
plt.figure(figsize=(12,8))
sns.scatterplot(x='high_cloud_cover_high_cld_lay', y='generated_power_kw', data=df, alpha = 0.5, label = 'high cloud cover')
sns.scatterplot(x='medium_cloud_cover_mid_cld_lay', y='generated_power_kw', data=df, alpha = 0.5 , label = 'Medium cloud cover')
sns.scatterplot(x='low_cloud_cover_low_cld_lay', y='generated_power_kw', data=df, alpha = 0.5,label = 'low cloud cover ')
plt.title('Generated Power vs Cloud Cover')
plt.xlabel('Cloud cover')
plt.ylabel('Generated power')
plt.legend()
plt.show()
```
