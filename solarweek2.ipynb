```python
def remove_outlier(col_name):
    # Calculate first and third quartiles
    Q1 = col_name.quantile(0.25)
    Q3 = col_name.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound
```

```python
lower, high = remove_outlier(df['mean_sea_level_pressure_MSL'])
```

```python
df['mean_sea_level_pressure_MSL'] = np.where(df['mean_sea_level_pressure_MSL']>high, high, df['mean_sea_level_pressure_MSL'])
```

```python
df['mean_sea_level_pressure_MSL'] = np.where(df['mean_sea_level_pressure_MSL']<lower, lower, df['mean_sea_level_pressure_MSL'])
```
